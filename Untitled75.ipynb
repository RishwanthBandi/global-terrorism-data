```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

np.random.seed(42)
n_samples = 1000

years = np.random.randint(1990, 2024, size=n_samples)
countries = np.random.choice(['USA', 'Iraq', 'Afghanistan', 'India', 'Pakistan', 'UK', 'France', 'Nigeria', 'Syria', 'Somalia'], size=n_samples)
attack_types = np.random.choice(['Bombing', 'Assassination', 'Armed Assault', 'Hijacking', 'Kidnapping'], size=n_samples)
weapons = np.random.choice(['Explosives', 'Firearms', 'Melee', 'Chemical'], size=n_samples)
casualties = np.random.poisson(lam=5, size=n_samples)
injuries = np.random.poisson(lam=10, size=n_samples)
success = np.random.choice([0, 1], size=n_samples)
property_damage = np.random.choice([0, 1], size=n_samples)


df = pd.DataFrame({
    'Year': years,
    'Country': countries,
    'Attack Type': attack_types,
    'Weapon Type': weapons,
    'Casualties': casualties,
    'Injuries': injuries,
    'Successful': success,
    'Property Damage': property_damage
})

print(df.head())
```

```python

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

print("Missing values:\n", df.isnull().sum())


for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col] = df[col].fillna(df[col].median())


for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled.")
```

```python

print("Mean Casualties:", df['Casualties'].mean())
print("Median Casualties:", df['Casualties'].median())
print("Std Dev:", df['Casualties'].std())


top_country = df.groupby('Country')['Casualties'].sum().idxmax()
print("Most Affected Country:", top_country)


top_10_countries = df.groupby('Country')['Casualties'].sum().sort_values(ascending=False).head(10)
print(top_10_countries)
```

```python
plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='barh', color=sns.color_palette("viridis", len(top_10_countries)))
plt.title("Top 10 Countries by Casualties")
plt.xlabel("Casualties")
plt.ylabel("Country")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Attack Type', data=df, order=df['Attack Type'].value_counts().index, palette='coolwarm')
plt.title("Attack Type Frequency")
plt.xticks(rotation=45)
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Casualties'], bins=50, color='teal', alpha=0.7)
plt.xlabel("Casualties")
plt.ylabel("Frequency")
plt.title("Distribution of Casualties")
plt.show()
```

```python
df['Weapon Type'].value_counts().head(5).plot.pie(autopct='%1.1f%%', figsize=(8, 8), cmap='Pastel1')
plt.title("Weapon Type Distribution")
plt.ylabel('')
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['number'])

plt.figure(figsize=(12, 6))
sns.heatmap(df_numeric.corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Terrorism Data")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby("Year").size().plot(kind="line", marker="o", color="blue")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.title("Terrorist Attacks Over Time")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Casualties"], y=df["Injuries"], alpha=0.5, color='darkred')
plt.xlabel("Casualties")
plt.ylabel("Injuries")
plt.title("Casualties vs Injuries")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Successful', y='Casualties', data=df, palette='Set2')
plt.title("Success vs Casualties")
plt.xlabel("Attack Successful")
plt.ylabel("Casualties")
plt.show()
```
